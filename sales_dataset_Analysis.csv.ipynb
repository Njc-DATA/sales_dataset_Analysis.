## 1. What are the total sales over time (daily/weekly/monthly)?

```python
# Convert 'Date' to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Sales over time
sales_over_time = data.groupby('Date')['Total_Sales'].sum().reset_index()
sales_over_time.head()
```

```python
visualize the result
```

```python
plt.figure(figsize=(12, 6))
plt.plot(sales_over_time['Date'], sales_over_time['Total_Sales'])
plt.title('Total Sales Over Time (Daily)')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

**total sales over time(daily) Insights**  

The overall sales over time(daily) shows interesting phenomena:

1. **Highest Sales Day**: The highest sales on January 1, 2025 crossed an amount of $12,790.29. This figure was most likely attained due to New Year sales or other holiday promotions.
  
2. **Sales Fluctuation**: The sales for the five days were highly inconsistent, the lowest sales were made on January 2, 2025 with an amount of only $1,228.11. This raises questions that need to be addressed.  

3. **Significant Sales Days**: We also observe that on January 4th, 2025, the sales reached a remarkable value of $10,540.43, showing us that there are some strategies that can be leveraged to reproduce this achievement.


### **Total Sales Over Time(Weekely)**

```python
weekly_sales = data.resample('W', on='Date')['Total_Sales'].sum().reset_index()
weekly_sales.head()
```

### **Visualize Result**

```python
plt.figure(figsize=(12, 6))
plt.plot(weekly_sales['Date'], weekly_sales['Total_Sales'])
plt.title('Total Sales Over Time (Weekly)')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

**Weekly Sales Insights**

The weekly sales data reveals a notable trend:

 **Increasing Sales Trend**:  Sales have been consistently increasing over the weeks, with the highest sales recorded in the week of February 2nd, 2025 ($55,592.51).

**Week-over-Week Growth**:  The sales growth from January 26th to February 2nd is particularly significant, with  an increase of $16,159.25 ($55,592.51  $39,433.26).

 **Strongest Week**: The week of February 2nd, 2025, is the strongest week in terms of sales, potentially indicating a successful marketing campaign, new product launch, or seasonal demand.

## **Total Sales Over Time(Monthly)**

```python
Monthly_sales = data.resample('ME', on='Date')['Total_Sales'].sum().reset_index()
Monthly_sales.head()
```

## **visualize Result**

```python
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Date'], monthly_sales['Total_Sales'])
plt.title('Total Sales Over Time (Monthly)')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

**Monthly Sales Insights**

The monthly sales data reveals:

**Peak Sales Month**: February 2025 had the highest sales ($189,451.48), potentially indicating strong demand or successful promotions.

**Decline in Recent Month**: May 2025 has significantly lower sales ($19,702.05) compared to previous months. This could be due to various factors such as seasonality, changes in market conditions, or incomplete data.

## **General view of the above plots**

The plots show sales trends over daily, weekly, and monthly periods.  Monthly and weekly views smooth out daily fluctuations, revealing clearer patterns.  Look for upward or downward trends, seasonality, and significant peaks or troughs.

## 2. Which region generates the most sales?

```python
sales_by_region = data.groupby('Region')['Total_Sales'].sum().reset_index()
top_region = sales_by_region.sort_values(by='Total_Sales', ascending=False).head(4)
print("\nTop Region by Sales:")
print(top_region)
```

### **visualize Result**

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Total_Sales', data=sales_by_region)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

**Top Region by Sales Insights**

The regional sales data shows:

**Top-Performing Regions**: The North and West regions are the top performers, with sales of 198,734.25 and 198,478.05, respectively.

**Close Competition**: The sales figures for the North and West regions are very close, indicating a competitive market.


**Lagging Regions**: The East region has the lowest sales ($157,134.77), suggesting opportunities for growth and improvement.



### **3. What is the average order value per region?**

```python
avg_order_value_per_region = (
    data.groupby('Region')['Total_Sales'].mean().reset_index()
)  # Assuming each row is an order
print("\nAverage Order Value per Region:")
print(avg_order_value_per_region)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Total_Sales', data=avg_order_value_per_region)
plt.title('Average Order Value per Region')
plt.xlabel('Region')
plt.ylabel('Average Order Value')
plt.show()
```

**Average Order Value (AOV) Insights**

The AOV data reveals:

 **Highest AOV**: The North region has the highest AOV (1,615.73), indicating potentially high-value customers or premium products.

 **Regional Variations**: AOV varies across regions, with the West (1,438.25) and South (1,409.46) regions having relatively lower AOVs compared to the North.
 
**Opportunities for Growth**: The East region's AOV ($1,378.38) is the lowest, suggesting potential for increasing average order values.


### **4. Who are the top 10 customers by total sales?**

```python
if 'Customer_Name' in data.columns:
    top_customers = (
        data.groupby('Customer_Name')['Total_Sales']
        .sum()
        .sort_values(ascending=False)
        .head(10)
        .reset_index()
    )
    print("\nTop 10 Customers by Sales:")
    print(top_customers)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Customer_Name', y='Total_Sales', data=top_customers)
plt.title('Top 10 Customers by Sales')
plt.xlabel('Customer Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

**Insights from Top 10 Customers by Sales**

The top 10 customers by sales are driving a significant portion of your business.

 **Top Customer**: Michael Soto is the top customer with total sales of 4,825.90, closely followed by Nicole Winters with  4,819.90.

 **Sales Distribution**: The top 3 customers (Michael Soto, Nicole Winters, and Richard Ellis) account for a substantial portion of total sales, indicating a concentrated customer base.
 
 **Customer Loyalty**: Richard Ellis and Richard Farrell are both top customers, suggesting potential opportunities for loyalty programs or personalized marketing.
 
 **Revenue Streams**: The sales figures indicate a range of customer spending habits, with the highest-spending customer generating $837.01 more than the lowest-spending customer in the top 10.

### 12. What payment methods are most commonly used?

```python
payment_method_counts = data['Payment_Method'].value_counts().reset_index()
payment_method_counts.columns = ['Payment_Method', 'Count']
print("\nPayment Method Counts:")
print(payment_method_counts)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Payment_Method', y='Count', data=payment_method_counts)
plt.title('Payment Method Usage')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### Insights from Payment Method Counts
The data shows:

**Most Popular Payment Method**: Debit Card (145) is the most frequently used payment method.

**Digital Payment Preference**: Debit Card, PayPal, and Credit Card account for 380 transactions, indicating a preference for digital payments. 

**Cash Usage**: Cash is still a significant payment method, with 120 transactions.



### 13. Is there any pattern between payment method and order value?
